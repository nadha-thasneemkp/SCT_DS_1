```python
import pandas as pd
df=pd.read_csv("/content/adult.csv")
print(df.head())
print("\n")
#no. of rows before cleaning
print(df.shape[0])
```

```python
#checking for missing values
df.replace('?',pd.NA,inplace=True)
before_cleaning=df.isnull().sum()
print(before_cleaning)

#handling missing value
df.dropna(inplace=True)
after_cleaning=df.isnull().sum()
print("\n")
print(after_cleaning)
print("\n")

#no. of rows after cleaning
print(df.shape[0])


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

#color for income groups
color = {'<=50K': 'orange', '>50K': 'green'}
plt.figure(figsize=(10,6))
sns.histplot(data=df,x="hours-per-week",hue="income",stat="density",bins=30,kde=True,element="step",palette=color,alpha=0.4)

#add legend
legend_elements =[
    Patch(facecolor='orange',alpha=0.4,label='<=50K - Histogram'),
    Line2D([0],[0],color='orange',lw=2,label='<=50K - KDE'),
    Patch(facecolor='green',alpha=0.4,label='>50K - Histogram'),
    Line2D([0],[0],color='green',lw=2,label='>50K - KDE'),
]
plt.legend(handles=legend_elements,title='Legend')
plt.tight_layout()
plt.grid(color='grey', linestyle='--', linewidth=1,alpha=0.4)
plt.title("Work Hours per Week by Income Group",fontsize=16)
plt.xlabel("Hours per Week",fontsize=14)
plt.ylabel("Density",fontsize=14)
plt.show()
```
